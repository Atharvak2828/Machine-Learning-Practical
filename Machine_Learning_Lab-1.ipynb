```python
# Filter for Stephen Curry's points, dropping any potential NaN values
curry_points = df[df['Player'] == 'Stephen Curry']['PTS'].dropna()

# Check if there's any data for Stephen Curry before plotting
if not curry_points.empty:
    plt.figure(figsize=(12, 7)) # Adjust figure size for more detail

    # Create the histogram with a Kernel Density Estimate (KDE) overlay
    sns.histplot(curry_points, kde=True,
                 bins=range(0, int(curry_points.max()) + 5, 2),
                 color='skyblue',
                 edgecolor='black',
                 alpha=0.7,
                 stat='density', # Use 'density' to normalize histogram bars, making KDE more intuitive
                 label='Points Distribution')

    # Add title and labels
    plt.title('Distribution of Stephen Curry\'s Points Scored (Bell Curve) with Outlier Analysis',
              fontsize=16, fontweight='bold', pad=20)
    plt.xlabel('Points Scored', fontsize=12)
    plt.ylabel('Density', fontsize=12)

    # Calculate Q1, Q3, and IQR
    Q1 = curry_points.quantile(0.25)
    Q3 = curry_points.quantile(0.75)
    IQR = Q3 - Q1

    # Define outlier bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = curry_points[(curry_points < lower_bound) | (curry_points > upper_bound)]

    # Add vertical lines for mean, median, and outlier bounds
    mean_pts = curry_points.mean()
    plt.axvline(mean_pts, color='red', linestyle='--', linewidth=2,
                label=f'Mean: {mean_pts:.2f} PTS')

    median_pts = curry_points.median()
    plt.axvline(median_pts, color='green', linestyle=':', linewidth=2,
                label=f'Median: {median_pts:.2f} PTS')

    plt.axvline(lower_bound, color='purple', linestyle='-.', linewidth=1.5,
                label=f'Lower Outlier Bound: {lower_bound:.2f}')
    plt.axvline(upper_bound, color='orange', linestyle='-.', linewidth=1.5,
                label=f'Upper Outlier Bound: {upper_bound:.2f}')

    # Highlight outliers on the plot if any exist
    if not outliers.empty:
        # Get the y-axis limit to place outlier markers just above the plot
        y_max = plt.gca().get_ylim()[1]
        # Plot outliers as distinct markers
        plt.scatter(outliers, [y_max * 0.95] * len(outliers), # Place markers near the top
                    color='black', marker='X', s=100, zorder=5,
                    label='Outlier(s)')
        # Optionally, add text labels for outliers
        for pt in outliers:
            plt.text(pt, y_max * 0.98, f'{pt:.0f}',
                     horizontalalignment='center', fontsize=9, color='black')
        print(f"\nOutliers detected: {outliers.tolist()}")
    else:
        print("\nNo outliers detected in Stephen Curry's points scored.")


    # Display the legend
    plt.legend(fontsize=10, bbox_to_anchor=(1.05, 1), loc='upper left') # Place legend outside

    # Add a grid for easier reading of values
    plt.grid(axis='y', alpha=0.75, linestyle='--')
    plt.grid(axis='x', alpha=0.75, linestyle='--')

    # Ensure layout is tight to prevent labels from overlapping, especially with external legend
    plt.tight_layout(rect=[0, 0, 0.85, 1]) # Adjust rect to make space for legend

    # Show the plot
    plt.show()
else:
    print("No data found for Stephen Curry's points to create the bell curve.")
    print("Please ensure 'Stephen Curry' exists in the 'Player' column of your dataset.")
```
